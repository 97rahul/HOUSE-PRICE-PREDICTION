# 1.DATA ANALYSIS

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

pd.pandas.set_option('display.max_columns',None)
```

```python
dataset = pd.read_csv('train.csv')
dataset.shape
```

```python
dataset.head()
```

# 1.1 MISSING VALUES

```python
features_with_na = [features for features in dataset.columns if dataset[features].isnull().sum()>1]

for feature in features_with_na:
    print(feature, round(dataset[feature].isnull().mean(),4), '% missing values')
```

```python
for feature in features_with_na:
    
    data = dataset.copy()
    
    data[feature] = np.where(data[feature].isnull(),1,0)
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

Clearly nan is affecting SalePrice.Hence,nan should be taken care in Feature Engineering.

```python
print('The length of Ids are {}.'.format(len(dataset.Id)))
```

# 1.2 NUMERICAL VARIABLES

```python
numerical_variables =[features for features in dataset.columns if dataset[features].dtypes !='O']

print(len(numerical_variables))

dataset[numerical_variables].head()
```

1.2.1 Temporal Variables

```python
year_features = [features for features in numerical_variables if 'Yr' in features or 'Year' in features]

year_features
```

```python
for feature in year_features:
    print(feature, dataset[feature].unique())
```

```python
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.title('House Price vs YearSold')
plt.xlabel('YearSold')
plt.ylabel('SalePrice')
```

```python
year_features
```

```python
for feature in year_features:
    if feature!='YrSold':
        data = dataset.copy()
        
        data[feature] = data['YrSold']-data[feature]
        
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

1.2.2 Discrete Variables

```python
discrete_variables = [features for features in numerical_variables if len(dataset[features].unique())<25 and features not in year_features + ['Id']]

len(discrete_variables)
```

```python
discrete_variables
```

```python
dataset[discrete_variables].head()
```

```python
for feature in discrete_variables:
    data = dataset.copy()
    
    data.groupby(feature)['SalePrice'].median().plot.bar(color=list('rgbkymc'))
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
    
```

1.2.3 CONTINUOUS VARIABLE

```python
continuous_variables = [features for features in numerical_variables if features not in discrete_variables + year_features + ['Id']]

len(continuous_variables)
```

```python
continuous_variables
```

```python
dataset[continuous_variables].head()
```

```python
for feature in continuous_variables:
    data = dataset.copy()
    
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(feature)
    plt.show()
```

```python
for feature in continuous_variables:
    data = dataset.copy()
    
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data['SalePrice'] = np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature)
        plt.show()
```

1.2.4 OUTLIERS

```python
for feature in continuous_variables:
    data = dataset.copy()
    
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```
